# Income by Zipcode

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
complete_file_zip_income = pd.read_csv("17zpallagi.csv")
```

```python
complete_file_zip_income.head()
```

```python
income_per_zipcode = complete_file_zip_income[["STATE", "zipcode", "A02650", "N1"]]

remove_invalid_zips_one_zero = income_per_zipcode.zipcode != 0
income_per_zipcode = income_per_zipcode[remove_invalid_zips_one_zero]

remove_invalid_zips_all_nines = income_per_zipcode.zipcode != 99999
income_per_zipcode = income_per_zipcode[remove_invalid_zips_all_nines]

income_per_zipcode = income_per_zipcode.rename(columns={"STATE": "state", "A02650": "total_income_amount", "N1": "total_returns_filed"})
```

```python
income_per_zipcode.head()
```

```python
incomes = income_per_zipcode.groupby(["zipcode", "state"]).sum()
```

```python
incomes["average_income"] = (incomes.total_income_amount / incomes.total_returns_filed) * 1000
```

```python
def format_deciamals(numbers):
    return float(f'{numbers:.2f}')
```

```python
incomes.average_income = incomes.average_income.apply(format_deciamals)
```
